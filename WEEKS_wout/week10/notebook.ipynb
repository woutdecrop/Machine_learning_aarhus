```python
def make_table(m, n):
    """Make a table with `m` rows and `n` columns filled with zeros."""
    return [[0] * n for _ in range(m)]
def compute_w(model, x):
    k = len(model.init_probs)
    n = len(x)
    print("k",k,"n",n)
    w = make_table(k, n)
    
    trans=model.trans_probs
    init=model.init_probs
    em=model.emission_probs
    # Base case: fill out w[i][0] for i = 0..k-1
    for i in range(k-1):
        w[i][0]=init[i]*em[i][0]
    
    # Inductive case: fill out w[i][j] for i = 0..k, j = 0..n-1
    for i in range(1,n):
        for j in range(0,k):
            z=[]
            for l in range(k):
                z.append(w[l][j])
            maxi=np.dot(z,trans[:][j])
 
            w[j][i]= np.max(maxi)*em[j][x[i]]
    return(w)
x=x[0:35]

w=compute_w(hmm_7_state,x)
#print(w[1,2])
b=np.asarray(w)
print(b[1,:])
print(b.shape)
opt_path_prob(b)
#print(np.max(w[:][14]))
def opt_path_prob(w):
    b=29
    l=7
    prob=1
    print("hallo")
    for i in range(b-1,0,-1):
        prob=prob*np.max(w[:,i])
        #prob= prob* np.max(w[:][i])
        print(prob)
    return prob
    pass

```
